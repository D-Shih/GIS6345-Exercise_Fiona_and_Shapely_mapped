```python
# env is geo_env
import pandas
import geopandas
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
crs = 'epsg:4326'
usa = geopandas.read_file("C:/Users/dshih/Documents/a_GIS6345/Week7/cb_2018_us_state_500k selection.shp")

usa = usa.to_crs(crs)
```

```python
filepoints = geopandas.read_file("some.shp") # type(filepoints) is geopandas.geodataframe.GeoDataFrame
filebuffer = geopandas.read_file("some_buffer.shp")
fileunion = geopandas.read_file("some_union.shp")
```

```python
filepoints = filepoints.set_crs(crs)
filebuffer = filebuffer.set_crs(crs)
fileunion = fileunion.set_crs(crs)
```

```python
ax = filepoints.plot(color='red', zorder=6)
ax.set_aspect('equal')
filebuffer.plot(ax=ax, color = 'green', edgecolor = 'black', zorder =3)
usa.plot(ax=ax, color = 'white', edgecolor = 'black', zorder = 1)
```

```python
ax = filepoints.plot(color='red', zorder=6)
ax.set_aspect('equal')
fileunion.plot(ax=ax, color = 'blue', edgecolor = 'black', zorder =2)
usa.plot(ax=ax, color = 'white', edgecolor = 'black', zorder = 1)
```
